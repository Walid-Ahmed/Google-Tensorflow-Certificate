```
#return X_train, X_test,word_index,tokenizer
def prepareNLPInputData(X_train,X_test,vocab_size=None,max_length=32,trunc_type='post',oov_tok = None):

  if (vocab_size is None) and (oov_token is None):
      tokenizer = Tokenizer( ) 
  else:
      #num_words: the maximum number of words to keep, based on word frequency. Only the most common `num_words-1` words will be kept. if not given all words will be included
      tokenizer = Tokenizer( num_words=vocab_size ,oov_token=oov_tok) #Only  the most frequent vocab_size  words are encoded
  
  tokenizer.fit_on_texts(X_train)
  word_index = tokenizer.word_index #only to view them later
  X_train = tokenizer.texts_to_sequences(X_train)  #turn your sentences into lists of values based on these tokens
  X_train = pad_sequences(X_train, padding='post', truncating=trunc_type, maxlen=max_length) #Padding return #<class 'numpy.ndarray'> is done by appropriate number of 0's before the sentence to match the length of the longest sentence or max length if provided

  X_test = tokenizer.texts_to_sequences(X_test)  #turn your sentences into lists of values based on these tokens
  X_test = pad_sequences(X_test, padding='post', truncating= trunc_type, maxlen=max_length) #Padding return #<class 'numpy.ndarray'> is done by a

  return X_train, X_test,word_index,tokenizer
```

```
def plot_graphs(history, key):
  plt.plot(history.history[key])
  plt.plot(history.history['val_'+key])
  plt.xlabel("Epochs")
  plt.ylabel(key)
  plt.legend([key, 'val_'+key])
  plt.show()
  
```

```
def readCSVFile(fileName,skipHeader=False):
  with open(fileName, 'r') as csvfile:
    reader = csv.reader(csvfile)
    lines=[]
    if(skipHeader):
      next(reader) # if needed to skip  first line
    for row in reader:  #row is a list
      lines.append(row)
  return lines # list of list, the inner list is the line tokenized by ,

```

```
def removestopWord(sentence,  stopwords):     
    for word in stopwords:   #removing stop words
        token = " " + word + " " #you do not want to remove substrings
        sentence = sentence.replace(token, " ")
        sentence = sentence.replace("  ", " ")
    return sentence    
```
